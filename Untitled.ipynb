```python
import os
import subprocess

def get_size(start_path):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(start_path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            if not os.path.islink(fp):
                total_size += os.path.getsize(fp)
    return total_size

def chunk_commit(folder_path, max_size_mb=100):
    total_size = 0
    max_size = max_size_mb * 256 * 256  # Convert MB to bytes

    files_to_commit = []
    batch_number = 1

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            file_size = os.path.getsize(file_path)
            
            if total_size + file_size > max_size:
                # Commit and push the current batch
                print(f"Committing and pushing batch {batch_number}...")
                commit_and_push_files(files_to_commit, batch_number)
                files_to_commit = []
                total_size = 0
                batch_number += 1

            files_to_commit.append(file_path)
            total_size += file_size

    # Commit and push any remaining files
    if files_to_commit:
        print(f"Committing and pushing final batch {batch_number}...")
        commit_and_push_files(files_to_commit, batch_number)

def commit_and_push_files(file_list, batch_number):
    # Add files to staging
    for file in file_list:
        subprocess.run(["git", "add", file], check=True)
        print(f"Added {file}")

    # Commit the staged files
    commit_message = f"Add batch {batch_number} of image files"
    subprocess.run(["git", "commit", "-m", commit_message], check=True)
    print(f"Committed batch {batch_number}.")

    # Push the commit
    subprocess.run(["git", "push"], check=True)
    print(f"Pushed batch {batch_number}.")

folder_path = 'static'  # Replace with the path to your images folder
chunk_commit(folder_path)
```

```python
import os
import subprocess

def chunk_commit_and_push(folder_path, files_per_chunk=100):
    files_to_commit = []
    batch_number = 1

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            files_to_commit.append(file_path)

            if len(files_to_commit) >= files_per_chunk:
                # Commit and push the current batch
                print(f"Committing and pushing batch {batch_number}...")
                commit_and_push_files(files_to_commit, batch_number)
                files_to_commit = []  # Reset for the next batch
                batch_number += 1

    # Commit and push any remaining files
    if files_to_commit:
        print(f"Committing and pushing final batch {batch_number}...")
        commit_and_push_files(files_to_commit, batch_number)

def commit_and_push_files(file_list, batch_number):
    # Add files to staging
    for file in file_list:
        subprocess.run(["git", "add", file], check=True)
        print(f"Added {file}")

    # Commit the staged files
    commit_message = f"Add batch {batch_number} of files"
    subprocess.run(["git", "commit", "-m", commit_message], check=True)
    print(f"Committed batch {batch_number}.")

    # Push the commit
    subprocess.run(["git", "push"], check=True)
    print(f"Pushed batch {batch_number}.")

folder_path = 'static'  # Replace with the path to your folder
chunk_commit_and_push(folder_path)
```
